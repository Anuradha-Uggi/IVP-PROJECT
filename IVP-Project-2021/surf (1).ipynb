```python
image1 = cv.imread(filename = '2651.png',
                   flags = cv.IMREAD_GRAYSCALE)
image2 = cv.imread(filename = 'affine_I.png',
                   flags = cv.IMREAD_GRAYSCALE)

import cv2 as cv
# Computing keypoints and descriptors using SURF.
surf = cv.xfeatures2d.SURF_create()
kp1, des1 = surf.detectAndCompute(image1,None)
kp2, des2 = surf.detectAndCompute(image2,None)
```

```python
# Brute force matching.
bf = cv.BFMatcher(cv.NORM_L1, crossCheck=True)
matches = bf.match(des1,des2)
```

```python
k1=len(kp1)
k2=len(kp2)
M=len(matches)
print('number of key points:',k1,'number of key points:',k2)
print('correct matches:',M)
print('%Match for i1:',M*100/k1)
print('%Match for i2:',M*100/k2)
```

```python

# Finding matches.
matches = sorted(matches, key = lambda x:x.distance)

# Drawing the matches between two images.
output = cv.drawMatches(img1 = image1,
                        keypoints1 = kp1,
                        img2 = image2,
                        keypoints2 = kp2,
                        matches1to2 = matches[:100],
                        outImg = None,
                        flags = cv.DrawMatchesFlags_NOT_DRAW_SINGLE_POINTS)

plt.imshow(output)
plt.show()
```
