```python
# Imports
import cv2 as cv
import matplotlib.pyplot as plt

# Open and convert the input and training-set image from BGR to GRAYSCALE
image1 = cv.imread(filename = '000002.jpg',
                   flags = cv.IMREAD_GRAYSCALE)

plt.subplot(1,2,1)
plt.imshow(image1)

# Open and convert the input and testing-set image from BGR to GRAYSCALE
image2 = cv.imread(filename = 'ther_trans.png',
                   flags = cv.IMREAD_GRAYSCALE)

plt.subplot(1,2,2)
plt.imshow(image2)
```

```python

# Finding keypoints and descriptors using ORB.

orb = cv.ORB_create()
kp1, des1 = orb.detectAndCompute(image1,None)
kp2, des2 = orb.detectAndCompute(image2,None)

# Brute force matching 
bf = cv.BFMatcher(cv.NORM_L1, crossCheck=True)
matches = bf.match(des1,des2)


k1=len(kp1)
k2=len(kp2)
M=len(matches)
print('number of key points:',k1,'number of key points:',k2)
print('correct matches:',M)
print('%Match for i1:',M*100/k1)
print('%Match for i2:',M*100/k2)

# Finding the matches.
matches = sorted(matches, key = lambda x:x.distance)

# Drawing the matches between two images.
output = cv.drawMatches(img1 = image1,
                        keypoints1 = kp1,
                        img2 = image2,
                        keypoints2 = kp2,
                        matches1to2 = matches[:100],
                        outImg = None,
                        flags = cv.DrawMatchesFlags_NOT_DRAW_SINGLE_POINTS)

plt.imshow(output)
plt.show()
```
